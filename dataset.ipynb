```python
from datasets import load_dataset
import os

# Load the dataset from Hugging Face
ds = load_dataset("Samarth0710/bharatanatyam-mudra-dataset")

# Choose your output directory
output_dir = "bharatanatyam_mudra_dataset"
os.makedirs(output_dir, exist_ok=True)

# Function to save images to class folders
def save_split(dataset_split, split_name):
    split_dir = os.path.join(output_dir, split_name)
    for idx, example in enumerate(dataset_split):
        label = example["label"]
        class_dir = os.path.join(split_dir, label)
        os.makedirs(class_dir, exist_ok=True)
        image = example["image"]
        # Save as jpg
        image_path = os.path.join(class_dir, f"{idx}.jpg")
        image.save(image_path)

# Save train split
save_split(ds["train"], "train")
# Save valid split (if exists; otherwise, you can create one with train_test_split)
if "validation" in ds:
    save_split(ds["validation"], "valid")
elif "test" in ds:
    save_split(ds["test"], "valid")  # Sometimes test split is used for validation

print("Done!")
```

```python
import os
import shutil
import random
from pathlib import Path

base_dir = Path(r"E:\sih\bharatanatyam_mudra_dataset\train")
train_dir = Path(r"E:\sih\bharatanatyam_mudra_dataset\train_split")
val_dir = Path(r"E:\sih\bharatanatyam_mudra_dataset\val")

split_ratio = 0.8  # 80% train, 20% validation

for cls in os.listdir(base_dir):
    cls_path = base_dir / cls
    if not cls_path.is_dir():
        continue

    images = list(cls_path.glob("*.*"))
    random.shuffle(images)
    split_idx = int(len(images) * split_ratio)

    train_imgs = images[:split_idx]
    val_imgs = images[split_idx:]

    # Create class folders in output dirs
    (train_dir / cls).mkdir(parents=True, exist_ok=True)
    (val_dir / cls).mkdir(parents=True, exist_ok=True)

    # Copy files
    for img in train_imgs:
        shutil.copy(img, train_dir / cls)
    for img in val_imgs:
        shutil.copy(img, val_dir / cls)

print("Dataset split complete.")
```
